```python
from matplotlib import collections as mc
import matplotlib.pyplot as plt
import numpy as np
import gaussianprimes as gp
```

# Generating Gaussian primes and basic operations

Getting Gaussian primes up to norm $1000$ in the first quadrant. A `Gints` object is returned, which is an `np.array` with a few special attributes and methods. Printing the Gaussian primes as `np.array` of complex numbers.

```python
a = gp.gprimes(1000)
a.to_complex()
```

Counting Gaussian primes by powers of 10. Each prime in the first quadrant has four associates; each associate is counted separately.

```python
for k in range(10):
    print(f'Number of primes up to norm 10^{k}: {gp.count(10 ** k)}')
```

We can generate Gaussian primes within a specified sector in complex plane. Below we generate Gaussian primes with angle measure in the interval $[0.1, 0.6)$ and norm up to $10^5$. We plot them.

```python
s = gp.gprimes_sector(10 ** 5, .1, .6)
s.plot()
```

Looking at a rectangular block of Gaussian primes. The real and imaginary parts $a$ and $b$ satisfy $10000 \le a < 10200$ and $20000 \le b < 20100$.

```python
b = gp.gprimes_block(10000, 20000, 200, 100)
b.plot()
```

A `Gints` object includes several basic methods to explore the generated primes. Because the `Gints` class is derived from `np.array`, standard numpy tools, including slicing and transposing, are available. Below we get the norms and angles of Gaussian primes from the block above and print the first 5.

```python
print(b.norms()[:5])
print(b.angles()[:5])
```

The only interesting arithmetic operation built into the `Gints` class is a mod operation based on the division algorithm in the Gaussian integers. Below we calculate the remainder upon division by $4 + 2i$ for Gaussian primes up to norm 1000.

```python
a % (4 + 2j)
```

# Racing Primes in Sectors and Angular Distribution

Akin to classical prime number races, we could use the mod operation above to race Gaussian primes in residue classes. We can also race Gaussian primes in sectors. Below we count primes with up to norm $10^9$ in several sectors in the complex plane.

```python
for k in range(20):
    angle0 = k / 100
    angle1 = (k + 1) / 100
    count = gp.count_sector(10 ** 9, angle0, angle1)
    print('# of primes in sector [{:.2f}, {:.2f}) up to norm 10^9: {}'.format(angle0, angle1, count))
```

The `gintsieve` module contains the class `SectorRaceWrapper` which can be used to conduct races between two sectors. Below we race the primes in the sectors defined by the intervals $[0.1, 0.3)$ and $[0.3, 0.5)$. The two sectors must have the same central angle.

```python
s = gp.Race(10 ** 8, .1, .3, .3, .5, 10000)
```

Let $\pi(x; \alpha, \beta)$ denote the number of Gaussian primes up to norm $x$ with angle measure in the interval $[\alpha, \beta)$. Below the difference $\pi(x; \alpha, \beta) - \pi(x; \gamma, \delta)$ is plotted for 10000 intermediate values of up to $x$. The red curve represents the normalizing factor $\frac{\sqrt{x}}{\log x}$.

```python
s.plot_race()
```

Plotting a Shanks-inspired histogram showing the ratio of the blue graph to the red graph in the plot above. Because there is no log-scaling, this histogram is not expected to converge to a probability distribution as $x\to\infty$. It does not appear particularly convergy...

```python
s.plot_shanks()
```

Most importantly, we can calculate the logarithmic density of the leader. This number may indeed converge as $x\to\infty$.

```python
s.density()
```

 This race is fairly balanced. We can also visualize the primes we are racing and color them by sector.

```python
s.plot_sectors()
```

Now we show a prime number race with more bias. The following call may take a few minutes to run because the norm bound is so large. Decrease the number `10 ** 10` if feeling impatient.

```python
s = gp.Race(10 ** 10, 0, np.pi / 8, np.pi / 8, np.pi / 4, 100000)
s.plot_race()
print(s.density())
```

And another race in narrower sectors. Empirically there seems to be plenty of bias in these sector races. Note how narrow sectors lead to a less *bouncy* race.

```python
s = gp.Race(10 ** 10, .01, .02, .03, .04, 100000)
s.plot_race()
print(s.density())
```

The function `angular_dist(x, n)` plots a histogram of the distribution of the counts of Gaussian primes up to norm $x$ among $n$ equally-spaced sectors in the first quadrant. Some outliers (counts above the 99th percentile and below the 1st percentile) are omitted. As an example of such an outlier, the sector containing the positive real-axis also contains every inert primes $p\equiv 3\bmod 4$, thereby inflating its count. There is an optional flag to keep these outliers.

The histogram below looks normal, which gives supporting evidence that the error terms in prime counting functions over sectors vary normally across sectors. The analog in $\mathbb{Z}$ is the distribution of $\pi(x; n, a) - \frac1{\varphi(n)} \pi(x)$ as $a$ ranges over units mod $n$.

```python
gp.angular_dist(10 ** 9, 10000, ignore_outliers=True);
```

## The Gaussian Moat Problem

Below we show several functions to explore the Gaussian prime graph with a specified jump size. See the [readme](../README.md#the-gaussian-moat-problem) for a thorough discussion. First we plot the main connected component of the Gaussian prime graph for several different jump sizes. Note that there are discrete thresholds at which the adjacency condition in the graph becomes weaker. At these thresholds, each Gaussian integer has more neighbors. These thresholds occur at $\sqrt{2}, 2, \sqrt{8}, \sqrt{10}, 4, \sqrt{18}, \sqrt{20}, \dots$. We show the main component of the prime graph for several of these thresholds.

```python
jump_size = 2 ** 0.5
m = gp.moat_component(jump_size)
m.plot()
```

```python
jump_size = 2.0
m = gp.moat_component(jump_size)
m.plot()
```

```python
jump_size = 8 ** 0.5
m = gp.moat_component(jump_size)
m.plot()
```

```python
jump_size = 10 ** 0.5
m = gp.moat_component(jump_size)
m.plot()
```

```python
jump_size = 4
m = gp.moat_component(jump_size)
m.plot()
```

```python
jump_size = 18 ** 0.5
m = gp.moat_component(jump_size)
m.plot()
```

Going for the jump size $\sqrt{20}$ requires more memory than my laptop can handle.

Now we look at all components in the first octant up to a specified norm. We give distinct components random colors. The following cell takes a minute to run.

```python
components = gp.moat_components_to_norm(10 ** 0.5, 3000000)
fig, ax = plt.subplots(figsize=(12, 8))
for component in components:
    x = component[0]
    y = component[1]
    color = np.random.rand(3,)
    ax.plot(x, y, color=color, linestyle='', marker='o', markersize=1)
```

We can also generate components within a specified block.

```python
jump_size = 4

# Setting optional `ignore_edges` flag to False in order to get edge data
components, segments = gp.moat_components_in_block(jump_size, 10000, 5000, 300, 200, False)
fig, ax = plt.subplots(figsize=(12, 8))
for component in components:
    x = component[0]
    y = component[1]
    color = np.random.rand(3,)
    ax.plot(x, y, color=color, linestyle='', marker='o', markersize=3)
    
lc = mc.LineCollection(segments, color='black', linewidth=1)
ax.add_collection(lc)
plt.show()
```

```python
jump_size = 18 ** 0.5

# Setting optional ignore_edges flag to False in order to get edge data
components, segments = gp.moat_components_in_block(jump_size, 10000, 5000, 300, 200, False)
fig, ax = plt.subplots(figsize=(12, 8))
for component in components:
    x = component[0]
    y = component[1]
    color = np.random.rand(3,)
    ax.plot(x, y, color=color, linestyle='', marker='o', markersize=3)
    
lc = mc.LineCollection(segments, color='black', linewidth=1)
ax.add_collection(lc)
plt.show()
```

```python
jump_size = 20 ** 0.5

# Not getting edges here since the component is obvious
components = gp.moat_components_in_block(jump_size, 10000, 5000, 300, 200, True)
fig, ax = plt.subplots(figsize=(12, 8))
for component in components:
    x = component[0]
    y = component[1]
    color = np.random.rand(3,)
    ax.plot(x, y, color=color, linestyle='', marker='o', markersize=3)
    
plt.show()
```
